```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

Data Preprocessing 

```python
df = pd.read_csv('AAPL.csv')
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
df.head()
```

```python
df = df.loc[:,['Close']]
df
```

```python
plt.plot(df.index,df['Close'])
plt.title('Closing price trends')
plt.xlabel('Date')
plt.ylabel('Close')
plt.show()
```

```python
df['Actual_Returns'] = df['Close'].pct_change() #Calculates the rate of change from the previous day sales
df
df = df.dropna()
```

```python
def shortEma(data):
    return data['Close'].ewm(span=5, adjust=False).mean()

def middleEma(data):
    return data['Close'].ewm(span=21, adjust=False).mean()

def longEma(data):
    return data['Close'].ewm(span=63, adjust=False).mean()
```

```python
df['Short'] = shortEma(df).copy()
df['Middle'] = middleEma(df).copy()
df['Long'] = longEma(df).copy()
```

```python
df = df.dropna()
df
```

```python
def buy_sell_function(data):
    buy_list, sell_list = [np.nan] * len(data), [np.nan] * len(data)
    flag_long, flag_short = False, False

    for i in range(0, len(data)):
        if (
            data['Short'][i] > data['Middle'][i] and
            data['Middle'][i] > data['Long'][i] and
            flag_long is False and flag_short is False
        ):
            buy_list[i] = data['Close'][i]
            sell_list[i] = np.nan
            flag_long = True
        elif flag_long and data['Short'][i] < data['Middle'][i]:
            sell_list[i] = data['Close'][i]
            buy_list[i] = np.nan
            flag_long = False
        if (
            data['Short'][i] < data['Middle'][i] and
            data['Middle'][i] < data['Long'][i] and
            flag_long is False and flag_short is False
        ):
            buy_list[i] = data['Close'][i]
            sell_list[i] = np.nan
            flag_short = True
        elif flag_short and data['Short'][i] > data['Middle'][i]:
            sell_list[i] = data['Close'][i]
            buy_list[i] = np.nan
            flag_short = False
    return buy_list, sell_list

df['Buy'],df['Sell'] = buy_sell_function(df)
```

```python
plt.figure(figsize=(12.2, 4.5))
plt.title('Visualizing the Buy/Sell singals', fontsize=18)
plt.plot(df['Close'], label='Close Price', color='blue', alpha=0.35)
plt.plot(df['Short'], label='Short/Fast EMA', color='red', alpha=0.35)
plt.plot(df['Middle'], label='Middle/Medium EMA', color='orange', alpha=0.35)
plt.plot(df['Long'], label='Long/Slow EMA', color='green', alpha=0.35)
plt.scatter(df.index, df['Buy'], color='green', marker='^', alpha=1)
plt.scatter(df.index, df['Sell'], color='red', marker='v', alpha=1)
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price', fontsize=18)
plt.legend()
```

Setting the target for the dependent variable (y) 

```python
df['Signal'] = 0
#When the actual returns are greater than or equal 0,generate to buy signal
df.loc[(df['Actual_Returns'] >= 0),'Signal'] = 1
df.loc[(df['Actual_Returns'] < 0),'Signal'] = 0
df.head()
```

```python
df['Strategy_Returns'] = df['Actual_Returns']*df['Signal'].shift()
df
```

```python
cumulative_returns = (1 + df['Strategy_Returns']).cumprod()
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df.index, cumulative_returns)
ax.set_title('Baseline Strategy')
ax.set_xlabel('Date')
ax.set_ylabel('Returns')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,test_size=0.3,random_state = 42)
x_test_indexes = x_test.index
x_train_columns = x_train.columns
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
model = LogisticRegression(random_state = 42)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
logreg_pred_df = pd.DataFrame(index=x_test_indexes)
logreg_pred_df['Model_Predictions'] = y_pred 
logreg_pred_df['Actual_Returns'] = df['Actual_Returns']
logreg_pred_df['Strategy_Returns'] = logreg_pred_df['Actual_Returns'] * logreg_pred_df['Model_Predictions']
logreg_pred_df
```

```python
print(classification_report(y_test,y_pred))
```

```python
pca = PCA(n_components=2,random_state = 42)
x_train_reduced = pca.fit_transform(x_train)
x_test_reduced = pca.fit_transform(x_test)
model = LogisticRegression(random_state = 42)
model.fit(x_train_reduced,y_train)
y_pred = model.predict(x_test_reduced)
print(classification_report(y_test,y_pred))
```

Hyperparameter tuning 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# Define the hyperparameter grid for each model
logistic_params = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'max_iter': [100, 200, 300]
}

svm_params = {
    'C': [0.1, 1, 10],
    'gamma': [0.001, 0.01, 0.1, 1]
}

decision_tree_params = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize the models with a consistent random_state where applicable
logistic = LogisticRegression(random_state=42)
svm = SVC(probability=True, random_state=42)
decision_tree = DecisionTreeClassifier(random_state=42)

# Initialize GridSearchCV for each model
grid_logistic = GridSearchCV(logistic, logistic_params, cv=5, scoring='accuracy')
grid_svm = GridSearchCV(svm, svm_params, cv=5, scoring='accuracy')
grid_decision_tree = GridSearchCV(decision_tree, decision_tree_params, cv=5, scoring='accuracy')

# Fit the models using GridSearchCV
grid_logistic.fit(x_train, y_train)
grid_svm.fit(x_train, y_train)
grid_decision_tree.fit(x_train, y_train)

# Extract the best models
best_logistic_model = grid_logistic.best_estimator_
best_svm_model = grid_svm.best_estimator_
best_decision_tree_model = grid_decision_tree.best_estimator_

# Define a function to calculate profit from strategy returns
def calculate_profit(strategy_returns_df):
    cumulative_returns = (1 + strategy_returns_df['Strategy_Returns']).cumprod()
    profit = cumulative_returns.iloc[-1] - 1
    return profit

# Apply the best models to the test data
logreg_pred_df['Model_Predictions'] = best_logistic_model.predict(x_test)
svm_pred_df['Model_Predictions'] = best_svm_model.predict(x_test)
dt_pred_df['Model_Predictions'] = best_decision_tree_model.predict(x_test)

# Calculate the strategy returns for each model
logreg_pred_df['Strategy_Returns'] = logreg_pred_df['Actual_Returns'] * logreg_pred_df['Model_Predictions'].shift()
svm_pred_df['Strategy_Returns'] = svm_pred_df['Actual_Returns'] * svm_pred_df['Model_Predictions'].shift()
dt_pred_df['Strategy_Returns'] = dt_pred_df['Actual_Returns'] * dt_pred_df['Model_Predictions'].shift()

# Calculate the profit for each model
logistic_profit = calculate_profit(logreg_pred_df)
svm_profit = calculate_profit(svm_pred_df)
decision_tree_profit = calculate_profit(dt_pred_df)

# Print the profit for each model
print(f"Profit/loss returns from the logistic regression model: {logistic_profit}")
print(f"Profit/loss returns from the SVM model: {svm_profit}")
print(f"Profit/loss returns from the decision tree model: {decision_tree_profit}")

# Identify the model with the highest profit
max_profit = max(logistic_profit, svm_profit, decision_tree_profit)
best_model = ('Logistic Regression', logistic_profit) if max_profit == logistic_profit else \
             ('SVM', svm_profit) if max_profit == svm_profit else \
             ('Decision Tree', decision_tree_profit)

print(f"The best model based on profit is: {best_model[0]} with a profit of {best_model[1]}")
```
